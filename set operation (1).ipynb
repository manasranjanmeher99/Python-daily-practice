# super set, sub set, disjoin

```python
a={1,2,3,4,5,6,7,8,9}
b={3,4,5,6,7,8}
```

```python
c={10,20,30,40}
```

```python
a.issuperset(b)
```

```python
a.issuperset(c)
```

```python
b.issubset(a)
```

```python
c.isdisjoint(a)
```

```python
b.isdisjoint(a)
```
