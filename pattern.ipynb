# Pattern

```python
# right angle tringle pattern
for i in range(1,6):
    print(" * "*i)
```

```python
# right angle tringle pattern
for i in range(5,0,-1):
    print(" * "*i)
```

```python
for i in range(1,6):   # pyramidal pattern
    print(' '*(5-i)+" * "*i)
```

```python
for i in range(1,6):   # pyramidal pattern
    print(' '*(5-i)+"*"*(2*i-1))
```

```python
  # reverse pyramidal pattern

for i in range(5,0,-1): 
    print(' '*(5-i)+"*"*(2*i-1))
```

```python
# triangle Pattern

for i in range(1,6):
    print(" * "*i)
for i in range(4,0,-1):
    print(" * "*i)
```

```python
# diamond partten

for i in range(1,6):  
    print(' '*(5-i)+"*"*(2*i-1))
for i in range(4,0,-1): 
    print(' '*(5-i)+"*"*(2*i-1))
```

```python
# Hollow Square partten

for i in range(5):
    for j in range(5):
        if i==0 or i==4 or j==0 or j==4:
            print(" * ", end="")
        else:
            print('   ',end="")
    print()
```

```python
# full square partten

for i in range(5):
    for j in range(5):
        print(" * ", end="")
    print()
```

```python
# right angle triangle number paartten

for i in range(1,6):
    for j in range(1,i+1):
        print(j,'', end="")
        
    print()
```

```python
# Inverted right angle triangle number paartten

for i in range(5,0,-1):
    for j in range(1,i+1):
        print(j,'', end="")
        
    print()
```

```python
# Floyd's triangle

num=1
for i in range(1,6):
    for j in range(1,i+1):
        print(num,'',end="")
        num+=1
    print()
```

```python
# Hollow Right Angle Triangle

for i in range(1,6):
    for j in range(1,i+1):
        if j==1 or j==i or i==5:
            print('* ', end="")
        else:
            print('  ',end="")
    print()
```

```python
# Hollow Pyramid Pattern

for i in range(1,6):
    for j in range(5-i):
        print(' ', end=" ")
    for j in range(2*i-1):
        if j==0 or j==2*i-2 or i==5:
            print('*',end=' ')
        else:
            print(' ', end=' ')
    print()
```

```python
# Hollow Diamond Pattern
n=5
for i in range(1,n+1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(2*i-1):
        if j==0 or j==2*i-2:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()

for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(2*i-1):
        if j==0 or j==2*i-2:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
    
```

```python
# Hallow Diamond number pattern

n=5
for i in range(1,n+1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(2*i-1):
        if j==0 or j==2*i-2:
            print(i,end=' ')
        else:
            print(' ',end=' ')
    print()

for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(2*i-1):
        if j==0 or j==2*i-2:
            print(i,end=' ')
        else:
            print(' ',end=' ')
    print()
    
```

```python
# Butterfly Pattern

n=5
for i in range(1,n+1):
    for j in range(i):
        print('*',end=' ')
    for j in range(2*(n-i)):
        print(' ',end=' ')
    for j in range(i):
        print('*', end=' ')
    print()
for i in range(n,0,-1):
    for j in range(i):
        print('*',end=' ')
    for j in range(2*(n-i)):
        print(' ',end=' ')
    for j in range(i):
        print('*', end=' ')
    print()
```
