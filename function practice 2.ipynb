```python
# Recursion function

def func2(n):
    if n<=0:
        print("Done!")
    else:
        print(n)
        func2(n-1)

func2(5)
```

```python
# factorial number
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
        
    
n=int(input("Enter a number: "))
factorial(n)
```

```python
# prime number

def prime(n):
    if n<=1:
        return False
    else:
        for i in range(2,n):
            if n%i==0:
                return False
        else:
            return True
         
n= int(input("Enter a number: "))
if prime(n):
    print(f"{n} is a prime number.")
else:
    print(f"{n} is not a prime number.")
```
