# Loops

```python
for i in range(5):
    print("data science")
```

```python
i=1
while i<=5:  # while loops
    print('data science')
    i= i+1
```

```python
i=1
while i<=5:  # while loops
    print('data science :',i)
    i= i+1
```

```python
i=5
while i>=1:  # while loops
    print('data science :',i)
    i= i-1
```

```python
i =1
while i<=5:
    print("Data science")
    j= 1
    while j<=4:
        print("Technology")
        j=j+1
    i=i+1
    print()
```

```python
i =1
while i<=5:
    print("Data science", end="")
    j= 1
    while j<=4:
        print(" technology", end="")
        j=j+1
    i=i+1
    print()
```

```python
i = 1
while i<=2:
    j=0
    while j<=2:
        print(i*j, end=" ")
        j+=1
    print()
    i+=1
```
