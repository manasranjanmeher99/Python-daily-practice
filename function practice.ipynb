```python
def fun8(greetings,*name):        # "Hello" is assigned to greeting, and the rest are collected in names.
    print(greetings,name[0])
    print(greetings,name[1])
    print(greetings,name[2])
fun8("Hello","manas","akash","satish")
```

```python
def fun8(greetings,*names):           #"Hello" is assigned to greeting, and the rest are collected in names.
    for name in names:
        print(greetings,name)
fun8("Hello","manas","akash","satish")
```

```python
def fun9(*add):
    sum=0
    for num in add:
        sum= sum+num
    return sum
print(fun9(1,2,3,4))
print(fun9(3,4))
print(fun9(11,46,89))
```

```python
# Finding the maximum value:

def fun10(*numbers):
    if len(numbers)==0:
        return "zero argument pass."
    max_num= numbers[0]
    for num in numbers:
        if max_num< num:
            max_num=num
    return max_num
print("maximum number is:",fun10(9,11,7,23,6,67))
```
